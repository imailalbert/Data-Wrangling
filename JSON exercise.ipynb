```python
import pandas as pd
import json
from pandas.io.json import json_normalize
```

```python
#create a pandas DataFrame and show first 5 rows to get a 'taste' of the DataFrame
projects = pd.read_json('data/world_bank_projects.json')
projects.head()
```

1. Find the 10 countries with most projects

```python
# list column names to choose the one we will count the countries on
list(projects)
```

```python
# 'value_counts' counts unique occurrences of 'countryname' values; 'head(10)' returns 10 with most counts
top10 = projects.countryname.value_counts().head(10)
top10
```

2. Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
# normalize 'mjteme_namecode' content from our json file into dataframe
data = json.load(open('data/world_bank_projects.json'))
major_themes = json_normalize(data, 'mjtheme_namecode')
major_themes.head()
```

```python
'''
Count unique values in the 'name' column and return the top 10.
However, this is not completely accurate as empty strings (missing names) were counted as well.
The final result of the major project themes will be given at the end of this notebook 
after the missing names will have been filled
'''
major_themes.name.value_counts().head(10)
```

3. Create a dataframe with the missing names filled in.

```python
#split the normalized dataframe into two: one with missing names (empty string) and another that has names
no_name = major_themes[major_themes.name == '']
w_name = major_themes[major_themes.name != '']
no_name
```

```python
# create copies of the two dataframes so we don't get "CopyWarning" message
no_name_copy = no_name.copy()
w_name_copy = w_name.copy()

'''
iterate through both dataframes, matching values in the 'code' columns, and assign values from 'name' column 
from the dataframe with names to the matching 'name' cell in the dataframe with no names
'''
for i in no_name_copy.code.index:
    for j in w_name_copy.code.index:
        if no_name_copy.code[i] == w_name_copy.code[j]:
            no_name_copy.name[i] = w_name_copy.name[j]
            break
        else:
            continue
            
```

```python
# concatenate both dataframes and print the result
frames = [w_name_copy, no_name_copy]
result = pd.concat(frames)
result
```

```python
# 10 major project themes
result.name.value_counts().head(10)
```
